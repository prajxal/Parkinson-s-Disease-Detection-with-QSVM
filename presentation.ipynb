## Conclusions

*   **Model Performance**: The QSVM performed competitively, achieving comparable AUC to classical SVM-RBF in lower dimensions (4 and 8). Random Forest showed consistent performance but struggled with the high dimensionality of the raw features before PCA.
*   **Dimensionality**: Lower PCA dimensions (4-8) generally yielded better stability for the QSVM, likely due to the limited number of qubits and the density of information.
*   **Recommendations**:
    *   **Feature Engineering**: Fine-tuning the ResNet18 backbone could yield more discriminative features.
    *   **Quantum Kernel**: Experimenting with different entanglement structures (e.g., circular) might capture more complex correlations.
    *   **Data**: Increasing the dataset size or using data augmentation would improve generalization.
